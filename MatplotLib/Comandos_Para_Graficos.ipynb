```python
#Imports
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Graf√≠co Simple 2D

```python
#Plot Cria um Graf√≠co 2D definido os [eixo x], [eixo y]
plt.plot([1,2,3,6],[1,2,3,4])
#Definido Custumiza√ß√£o para o graf√≠co
plt.title("Exemplo")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")

#Show √© usado para mostrar o graf√≠co
plt.show()
```

# Trabando gr√°ficos com arquivos

# Manipulando propriedades do gr√°fico

# Plotando 2 gr√°ficos sobre o mesmo eixo com legendas

```python
#Pegando a Figura para permitir manipula√ß√µes
fig = plt.figure()
#Trocando BackGround
bg = fig.patch
bg.set_facecolor('white')

x = [1,2,3,4,5]
y = [1,2,3,4,5]
x2 = [2,4,5,9,8]
y2 = [1,6,2,3,4]


#Plotando e Definindo Configura√ß√£o da linha
plt.plot(x,y, 'blue', linewidth=3.3,  linestyle='--', label='Data 1')
plt.plot(x2,y2, 'red',linewidth=2,  linestyle='-', label='Data 2') #Segunda Linha
#Legendas
plt.legend(loc='upper right')


#Setando Cores dos referenciais
plt.title("Elemplo", color='yellow')
plt.xlabel("Eixo X", color="Green")
plt.ylabel("Eixo Y", color="Green")

#Tick Params muda as pauzinhos nos n√∫meros
plt.tick_params(axis='x', colors='red')
plt.tick_params(axis='y', color='red')
```

# Plotando v√°rios gr√°ficos na mesma figura

```python
#Criando o figure onde vai ficar os graficos
figure = plt.figure()
rect = figure.patch
rect.set_facecolor('white')

EX1 = [1,5,2,5,6]
EY1 = [1,2,3,4,5]
EX2 = [7,5,6,3,2]
EY2 = [2,3,4,5,2]

#Sub Gr√°fico 1
ax1 = figure.add_subplot(2,2,1) #Adicionando o sub plot em uma matriz 2x2 na primera casa: (2,2,1)
ax1.plot(EX1,EY1, 'green', linestyle='--')

#Sub Gr√°fico 2
ax2 = figure.add_subplot(2,2,2)
ax2.plot(EX2,EY2, 'red',linestyle='-')

#Sub Gr√°fico 3
ax1 = figure.add_subplot(2,2,3)
ax1.plot(EX2,EY2, 'gray', linestyle='--')

#Sub Gr√°fico 4
ax2 = figure.add_subplot(2,2,4)
ax2.plot(EX1,EY1, 'yellow',linestyle='-')

```

# Plotando gr√°ficos de barra

```python
from pylab import *

cand = ['Sardinha', 'Alface', 'Feij√£o', 'Pizza', 'Batata Doce', 'Frango']
popularity = [-5,2,3,2,4,6]

#Barh gera grafico na vetical, bar na horizontal
barh(cand, popularity, align='center', color='orange')
#bar(cand, popularity , align='center', color='orange')

grid(True) #Grid cria Grades no Gr√°fico
xlabel('Indicie de Votos')
ylabel('Comidas')
title('Grafico de Coluna')

show()
```

# Criando Gr√°ficos de Pizza

```python
#Dados para o gr√°fico
x_list = [1,4,3,2]
label_list = ["Cobras","Cachoros","Gatos","Pass√°ros"]

plt.axis('equal') #Deixa o Grafico em outra vis√£o
#Fun√ß√£o pie para gerar gr√°ficos em pizza, autopct coloca as porcentages
plt.pie(x_list, labels=label_list, autopct='%1.1f%%')

plt.title("Animais Preferidos")

plt.show()
```

# Gr√°ficos de linha 3D

```python
#Importa√ß√µes
from mpl_toolkits.mplot3d import axes3d

ax = plt.axes(projection='3d')

# Data for a three-dimensional line
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'gray')

# Data for three-dimensional scattered points
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens');
```
